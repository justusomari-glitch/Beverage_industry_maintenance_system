### First model which is the use of Isolation Forest and One class SVM to find anomalies in the sensor data

```python
from sklearn.ensemble import IsolationForest
from sklearn.pipeline import Pipeline,FeatureUnion
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
from sklearn.compose import ColumnTransformer
from sklearn.decomposition import PCA
```

```python
df_readings.columns
```

```python
df.columns
```

```python
readings=['temperature', 'vibration','pressure', 'rotation_speed', 'torque', 'voltage', 'current',
          'power_consumption', 'humidity', 'noise_level', 'flow_rate',
       'product_temperature', 'cleaning_cycle_status', 'oil_level',
       'bearing_temperature', 'hours_since_maintenance']
```

```python
x_anomaly=df[readings]
```

```python
x_anomaly.head()
```

```python
x_anomaly.shape
```

#### begining of modeling

```python
processing=ColumnTransformer([
    ('std',StandardScaler(),readings)
])
```

```python
ir=IsolationForest(random_state=42)
```

```python
anomaly_pipeline=Pipeline([
    ('processing',processing),
    ('ir',ir)
])
```

```python
param={
    "ir__n_estimators":[100],
    "ir__contamination":[0.06],
    "ir__max_samples":[0.1],
    "ir__max_features":[0.5]
}
    
```

```python
def anomaly_scorer(estimator,X):
    X_transformed=estimator.named_steps['processing'].transform(X)
    return estimator.named_steps['ir'].score_samples(X_transformed).mean()
```

```python
anomaly_model=GridSearchCV(
    estimator=anomaly_pipeline,
    param_grid=param,
    cv=3,
    verbose=2,
    scoring=anomaly_scorer
)
```

```python
anomaly_model.fit(x_anomaly)
```

```python
anomaly_prediction=anomaly_model.predict(x_anomaly)
```

```python
anomaly_binary=np.where(anomaly_prediction==-1,1,0)
```

```python
x_table=x_anomaly.copy()
```

```python
x_table['anomaly']=anomaly_binary
```

```python
x_table['anomaly'].value_counts()
```

```python
df_readings.columns
```

```python
df_readings['status'].value_counts()
```

```python
### looking at the results and scores
```

```python
print("="*50)
print("Best Score:",anomaly_model.best_score_)
print("Best Parameters:",anomaly_model.best_params_)
```

```python
results=pd.DataFrame(anomaly_model.cv_results_)
results=results[['params','mean_test_score','std_test_score','rank_test_score']]
results=results.sort_values('rank_test_score')
print("\nFULL RESULTS TABLE:")
print(results)
```

```python
import joblib
```

```python
joblib.dump(anomaly_model,'anomaly.pkl',compress=3)
print("Model saved Succesfull")
```

## doing the modeling of our second model failure type prediction

```python
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier,VotingClassifier
from sklearn.model_selection import RandomizedSearchCV,train_test_split
from sklearn.pipeline import Pipeline,FeatureUnion
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.feature_selection import SelectKBest,f_classif
from sklearn.compose import ColumnTransformer
from sklearn.decomposition import PCA
from imblearn.over_sampling import SMOTE
from sklearn.metrics import classification_report,make_scorer,recall_score
```

```python
df['criticality_level'].value_counts()
```

```python
df.columns
```

```python
categs=['zone','criticality_level']
```

```python
numerics=[sensors]
```

```python
x_type=df[categs+sensors]
```

```python
y_type=df['failure_type']
```

```python
x_type.head()
```

```python
x_type.shape
```

```python
x_type.isnull().any()
```

```python
y_type.head()
```

```python
train_x_type,test_x_type,train_y_type,test_y_type=train_test_split(x_type,y_type,test_size=0.3,random_state=42)
```

```python
print(train_x_type.shape)
print(test_x_type.shape)
print(train_y_type.shape)
print(test_y_type.shape)
```

```python
type_process=ColumnTransformer([
    ('oht',OneHotEncoder(drop='first'),categs),
    ('std',StandardScaler(),sensors)
])
```

```python
type_features=FeatureUnion([
    ('skt',SelectKBest(score_func=f_classif))
])
```

```python
rf=RandomForestClassifier()
xg=XGBClassifier()
```

```python
ensemble=VotingClassifier(
    estimators=[('rf',rf),
                ('xg',xg)],
    voting='soft'
)
```

```python
type_pipeline=Pipeline([
    ('type_process',type_process),
    ('type_features',type_features),
    ('ensemble',ensemble)
])
```

```python
type_param={
    "ensemble__rf__n_estimators":[50,100],
    "ensemble__rf__max_depth":[10,30],
    "ensemble__rf__max_features":['sqrt','log2'],
    "ensemble__xg__n_estimators":[50,100],
    "ensemble__xg__max_depth":[3,10],
    "ensemble__xg__learning_rate":[0.01,0.05,0.3],
    "ensemble__rf__class_weight":['balanced','balanced_subsample']
}
```

```python
scorer=make_scorer(recall_score,average='macro')
```

```python
failure_type=RandomizedSearchCV(
    estimator=type_pipeline,
    param_distributions=type_param,
    cv=3,
    verbose=2,
    error_score='raise'
)
```

```python
failure_type.fit(train_x_type,train_y_type)
```

```python
type_preds=failure_type.predict(test_x_type)
```

```python
type_preds
```

```python
trial=test_x_type.copy()
```

```python
trial['condition']=type_preds
```

```python
trial['condition'].value_counts()
```

```python
print(classification_report(test_y_type,type_preds))
```

```python
import joblib
```

```python
joblib.dump(failure_type.best_estimator_,'failure_type.pkl',compress=3)
print("Model Saved")
```

```python
severe_ensemble=VotingClassifier(
    estimators=[('rf',rf),
                ('xg',xg)],
    voting='soft'
)
```

```python
severe_pipeline=ImbPipeline([
    ('severe_process',severe_process),
    ('variance',VarianceThreshold(threshold=0.0)),
    ('skt',SelectKBest(score_func=f_classif)),
    ('smote',SMOTE(random_state=42)),
    ('severe_ensemble',severe_ensemble)
])
```

```python
severe_param={
    'skt__k':[5,10,15],
    "severe_ensemble__rf__n_estimators":[100,200],
    "severe_ensemble__rf__max_depth":[20,30],
    "severe_ensemble__xg__n_estimators":[100,200],
    "severe_ensemble__xg__max_depth":[3,10],
    "severe_ensemble__xg__learning_rate":[0.05,0.1],
}
```

```python
severity_model=GridSearchCV(
    estimator=severe_pipeline,
    param_grid=severe_param,
    cv=3,
    verbose=2,
    error_score='raise'
)
```

```python
print("\nFitting Severity Model....")
severity_model.fit(train_x_severe,train_y_severe)
```

```python
print(classification_report(test_y_severe,severe_predictions))
```

## We now Move to the Decision Engine

```python
def decision_engine(row):
    ## prediction based decisions
    if(
        row['anomaly']==0 and
        row['failure_type']=='No Failure' and 
        row['root_cause']=='No Fault Detected' and
        row['corrective_action']=='No action needed'):
        return 'Normal Operations'
        ## Severity Based Decisions
    scores=row['mcdm_score']
    severity=row['severity']
    if row['anomaly']==1 and row['failure_type']=='No Failure':
        if severity== 'High':
            return "Urgent Inspection Required"
        if severity== 'Medium':
            return "Schedule Inspecton"
        if severity== 'Low':
            return "Monitor Closely"
        ## score based decisions
    if scores>=0.7:
        return "Critical Fault: Initiate Corrective Maintenance"
    if scores>=0.5:
        return "Fault Detected: Plan Maintenance Within 24 Hours"
    if scores>=0.3:
        return "Minor Fault : Schedule Preventive Maintenance"
    else:
        return "Normal Operations"
```

```python
results=pd.DataFrame({
    'anomaly':anomaly_binary,
    'severity':severity,
    'failure_type':failure,
    'root_cause':root_cause,
    'corrective_action':recommendation,
    'mcdm_score':scores,
    'priority':priority
})
```

```python
results['severity']=results.apply(
    lambda row: 'Low' if
    (row['anomaly']==0 and
     row['failure_type']=='No Failure' and
     row['root_cause']== 'No Fault Detected')
    else
    row['severity'],axis=1)
```

```python
results['priority']=results.apply(decision_engine,axis=1)
```
